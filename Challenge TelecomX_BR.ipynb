#üìå Extrac√£o

```python
import pandas as pd
import numpy as np
import requests
import json

url = "https://raw.githubusercontent.com/alura-cursos/challenge2-data-science/main/TelecomX_Data.json"

response = requests.get(url)
data = response.json()

df = pd.json_normalize(data)

print("\nüîç Primeiras linhas do DataFrame:")
display(df.head())

print("\nüìä Informa√ß√µes do DataFrame:")
display(df.info())

print("\nüìà Estat√≠sticas descritivas:")
display(df.describe(include='all'))
```

#üîß Transforma√ß√£o

```python
rename_columns = {
    "customerID": "id_cliente",
    "Churn": "cancelou",
    "customer.gender": "genero",
    "customer.SeniorCitizen": "idoso",
    "customer.Partner": "parceiro",
    "customer.Dependents": "dependentes",
    "customer.tenure": "tempo_contrato_meses",
    "phone.PhoneService": "telefone_servico",
    "phone.MultipleLines": "linhas_multiplas",
    "internet.InternetService": "internet_servico",
    "internet.OnlineSecurity": "seguranca_online",
    "internet.OnlineBackup": "backup_online",
    "internet.DeviceProtection": "protecao_dispositivo",
    "internet.TechSupport": "suporte_tecnico",
    "internet.StreamingTV": "tv_streaming",
    "internet.StreamingMovies": "filmes_streaming",
    "account.Contract": "tipo_contrato",
    "account.PaperlessBilling": "fatura_digital",
    "account.PaymentMethod": "metodo_pagamento",
    "account.Charges.Monthly": "custo_mensal",
    "account.Charges.Total": "custo_total"
}

df_br = df.rename(columns=rename_columns)

traducao_categorias = {
    "Yes": "Sim",
    "No": "N√£o",
    "No phone service": "Sem servi√ßo telef√¥nico",
    "No internet service": "Sem servi√ßo de internet",
    "Month-to-month": "Mensal",
    "One year": "Anual",
    "Two year": "Bienal",
    "Female": "Feminino",
    "Male": "Masculino",
    "Electronic check": "D√©bito Autom√°tico",
    "Mailed check": "Boleto Enviado",
    "Bank transfer (automatic)": "Transfer√™ncia banc√°ria (autom√°tica)",
    "Credit card (automatic)": "Cart√£o de cr√©dito (autom√°tico)",
    "DSL": "DSL",
    "Fiber optic": "Fibra √≥tica"
}

for col in df_br.select_dtypes(include='object').columns:
    if col != "id_cliente":
        df_br[col] = df_br[col].str.strip().replace(traducao_categorias)

df_br["cancelou"] = df_br["cancelou"].str.strip().str.capitalize()
df_br["cancelou"] = df_br["cancelou"].replace({
    "Yes": "Sim", "No": "N√£o",
    "": np.nan, "Nan": np.nan
})

df_br = df_br[df_br["cancelou"].isin(["Sim", "N√£o"])].copy()

df_br["custo_total"] = pd.to_numeric(df_br["custo_total"], errors='coerce')

df_br["idoso"] = df_br["idoso"].map({1: "Sim", 0: "N√£o"})

print("\nüîç Estrutura do DataFrame ap√≥s transforma√ß√£o:")
print(df_br.info())

print("\nüìä Valores √∫nicos em 'cancelou':")
print(df_br["cancelou"].value_counts())

print("\nüìä Amostra dos dados transformados:")
display(df_br.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="white", palette="Set2")

churn_palette = {"Sim": "#e74c3c", "N√£o": "#2ecc71"}
```

```python
plt.figure(figsize=(7, 5))
ax = sns.countplot(data=df_br, x='cancelou', hue='cancelou', palette=churn_palette, legend=False)
plt.title("Distribui√ß√£o de Cancelamentos (Churn)", fontsize=14)
plt.xlabel("Cancelou Servi√ßo")
plt.ylabel("N√∫mero de Clientes")

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + 0.1, p.get_height() + 20))

sns.despine()
plt.tight_layout()
plt.show()
```

```python
categorical_cols = ['genero', 'idoso', 'parceiro', 'dependentes', 'tipo_contrato', 'fatura_digital']
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    ax = sns.countplot(data=df_br, x=col, hue='cancelou', palette=churn_palette)
    plt.title(f"Cancelamento por {col.capitalize()}", fontsize=13)
    plt.xlabel(col.capitalize())
    plt.ylabel("N√∫mero de Clientes")
    plt.xticks(rotation=30)
    for container in ax.containers:
        ax.bar_label(container, fmt='%d', label_type='edge', padding=3)
    sns.despine()
    plt.tight_layout()
    plt.show()
```

```python
numeric_cols = ['tempo_contrato_meses', 'custo_mensal', 'custo_total']
nome_amigavel = {
    'tempo_contrato_meses': 'Tempo de Contrato (meses)',
    'custo_mensal': 'Custo Mensal (R$)',
    'custo_total': 'Custo Total (R$)'
}

for col in numeric_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(data=df_br, x=col, hue='cancelou', palette=churn_palette, kde=True, bins=30, alpha=0.6)
    plt.title(f"Distribui√ß√£o de {nome_amigavel[col]} por Churn", fontsize=13)
    plt.xlabel(nome_amigavel[col])
    plt.ylabel("Frequ√™ncia")
    sns.despine()
    plt.tight_layout()
    plt.show()
```

```python
sns.set_theme(style="whitegrid")

# 1. Countplot por tipo de contrato
plt.figure(figsize=(10, 5))
ax = sns.countplot(data=df_br, x='tipo_contrato', hue='cancelou', palette=churn_palette)
plt.title("Cancelamento por Tipo de Contrato", fontsize=14)
plt.xlabel("Tipo de Contrato")
plt.ylabel("N√∫mero de Clientes")
plt.xticks(rotation=15)
plt.legend(title="Cancelou")
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', padding=3)
sns.despine()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df_br, x='cancelou', y='tempo_contrato_meses', hue='cancelou', palette=churn_palette, width=0.6, fliersize=2, legend=False)
plt.title("Tempo de Contrato (meses) por Churn", fontsize=14)
plt.xlabel("Cancelou")
plt.ylabel("Tempo de Contrato (meses)")
sns.despine()
plt.tight_layout()
plt.show()
```

```python
for col in categorical_cols:
    porcentagem = (
        df_br
        .groupby(col)['cancelou']
        .value_counts(normalize=True)
        .unstack()
        .fillna(0)
        .sort_values(by="Sim", ascending=False)
        * 100
    )

    display(
        porcentagem
        .style
        .format("{:.1f}%")
        .background_gradient(cmap='Blues')
        .set_caption(f"Distribui√ß√£o percentual de cancelamento por '{col}'")
    )
```

## 1. üéØ **Foco no P√∫blico com Maior Risco de Cancelamento**  

### üîπ **Idosos**  
- **Churn alto:** 41,7% dos clientes idosos cancelam.  
- **A√ß√£o recomendada:** Criar planos personalizados para idosos, com:  
  - Atendimento priorit√°rio e humanizado  
  - Campanhas educativas sobre servi√ßos digitais

### üîπ **Clientes sem parceiro ou sem dependentes**  
- **Sem parceiro ‚Üí** 33,0% de churn  
- **Sem dependentes ‚Üí** 31,3% de churn  
- **Insight:** Clientes solteiros ou mais independentes s√£o mais vol√°teis.  
- **A√ß√£o recomendada:**  
  - Oferecer pacotes individuais mais flex√≠veis

### üîπ **Fatura Digital**  
- **Quem usa fatura digital cancela mais (33,6%)** vs. quem n√£o usa (16,3%).  
- **Poss√≠vel problema:** Comunica√ß√£o falha ou falta de transpar√™ncia no faturamento digital.  
- **A√ß√£o recomendada:**  
  - Melhorar clareza e usabilidade da fatura digital

  - Incluir resumo visual da cobran√ßa
